```python
x = device['current']
y = device['power']

#regression line
slope, intercept, r, p_value, std_err= stats.linregress(x,y)
line = slope*x + intercept

plt.plot(x,line,
         color='skyblue',
         label=f'regression line(r = {r:.3f}')

plt.scatter(x,y,
            color='#F25067',
            label='data points')
plt.title('Device Current and Power Response')
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.legend()
plt.show()
```

# Bar Graph

```python
bar = plt.bar (['Jaguar','Panther'],[jag_mean,pan_mean],color = ['pink','#299367'])
plt.title('Average Resistance by Machine')

#bar labels
plt.ylabel(['Resistance (ohm)'], color = '#099367')
plt.bar_label (bar)
plt.margins(y=0.1)
plt.show()
```
